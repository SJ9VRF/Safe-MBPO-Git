```python
path_cheetah = "/content/drive/My Drive/data-result/cheetah/episodes.csv"
humanoid = pd.read_csv(path_cheetah)

d_val = 12
mean_1 = humanoid['eval return mean'][d_val:]
mean_1 = [item-d_val+300 for item in mean_1]
std_1 = humanoid['eval return std'][d_val:]
x = humanoid['total violations'][d_val:]
x = [item-d_val+10 for item in x]

plt.plot(x, mean_1, 'g-', label='SMBPO')
plt.fill_between(x, mean_1 - std_1, mean_1 + std_1, color='g', alpha=0.2)

plt.xlim(0, 300)
plt.ylim(-400, 12000)

txt="cheetah-no-flip"
plt.figtext(0.5, 0.9, txt, wrap=True, horizontalalignment='center', fontsize=20)


plt.xlabel('Total violations', fontsize=18)
plt.ylabel('return', fontsize=16)

plt.legend()
plt.show()


plt.plot(x, mean_1, 'g-', label='SMBPO')
plt.fill_between(x, mean_1 - std_1, mean_1 + std_1, color='g', alpha=0.2)


txt="cheetah-no-flip"
plt.figtext(0.5, 0.9, txt, wrap=True, horizontalalignment='center', fontsize=20)


plt.xlabel('Total violations', fontsize=18)
plt.ylabel('return', fontsize=16)

plt.legend()
plt.show()
```
